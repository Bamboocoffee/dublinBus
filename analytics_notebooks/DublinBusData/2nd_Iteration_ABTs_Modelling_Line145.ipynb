## FUNCTION: LABEL ENCODER FOR STOPS

```python
def add_label_encoding(df, categoricals_to_encode):
    
    """
    Performs label-encoding of categorical features.
    Suitable for tree-based methods that do not 
    add ordinality meaning.
    
    INPUT: dataframe and list of features to encode
    
    OUTPUT: dataframe with new encoded features; 
        originals removed.
        
    USE: run on total data after all experiments 
    for map of all values > encodings.
    
    """
    
    # iterate through features list
    # label-encode their values
    
    for feature in categoricals_to_encode:
        le = LabelEncoder()
        name = feature + '_label'
        df[name] = le.fit_transform(df[feature])
        
        # create a mapping of original to labelled
        # values
        
        encoding = df[name].tolist()
        original = le.inverse_transform(df[name]).tolist()
        
        dictionary = dict(zip(original, encoding))
        
        # save the mapping to disk, as json
        
        file_path = '/Users/davidodwyer/Documents/studyCS/Semester_3/data/dataframes/label_encode_maps'
        dict_name = feature + '_label_encode_map.json'
        dir_file = os.path.join(file_path, dict_name)
        
        with open(dir_file, 'w') as json_file:
            json.dump(dictionary, json_file)
         
    # drop the original, unencoded features        
            
    for feature in categoricals_to_encode:
        df.drop(feature, axis=1, inplace=True)
        
    return df
        
```

## FUNCTIONS: GPS DISTANCE FEATURE

```python
def add_stop_lat_lon_features(df):
    
    """
    Adds latitude and longitude features for each 
    stop / row of data.
    """
    
    df['stop_lat'] = df.STOPPOINTID
    df['stop_lon'] = df.STOPPOINTID
    
    df = df.astype({
        'stop_lat':'int32',
        'stop_lon':'int32'
    })
    
    # for stops in dictionary and df, map; retain values
    # for stops not in dictionary
    
    df['stop_lat'] = df['stop_lat'].map(stops_lat).fillna(df['stop_lat'])
    df['stop_lon'] = df['stop_lon'].map(stops_lon).fillna(df['stop_lat'])
      
    # stops in the dublin but dataset and in the mapping dict
    stops_in_current_dataframe = list(set(df.STOPPOINTID.unique()).intersection(set(stops_latlon.keys())))
    
    # stops in dublin bus dataset but not in mapping dict
    stops_not_in_dict = list(set(df.STOPPOINTID.unique()).difference(set(stops_latlon.keys())))
    
    for stop in stops_not_in_dict:
        while stop not in stops_in_current_dataframe:
            stop -= 1
        df.loc[df.STOPPOINTID == stop, 'stop_lat'] = stops_latlon[stop][0]
        df.loc[df.STOPPOINTID == stop, 'stop_lon'] = stops_latlon[stop][1]
    
    return df
```

```python
def distance_from_city_centre(row):
    
    """
    Calculates the great circle distance of a lat/lon
    pair from the city centre defined as O' Connell 
    bridge, Dublin.
    
    Wrapped in def add_distance_from_centre_feature()
    """
    
    centre_gps = (53.342998628, -6.256165642)
    
    # create tuple of coordinates from latitude
    # and longitude of given row
    
    stop_gps = (row.stop_lat, row.stop_lon)
    
    # calculate 
    
    distance = haversine(centre_gps, stop_gps)
    
    return distance
    
```

```python
def add_distance_from_centre_feature(df):
    
    """
    Creates a new feature: distance of each row's stop from
    the city centre
    """
    
    df['distance_centre'] = df.apply(distance_from_city_centre, axis=1)
    
    return df
```

```python
def add_time_of_day_func1(df):
    
    """
    Adds two features, together representing the cyclic ordinal character
    of time. Time of day of expected time of arrival at stop.
    
    This is adapted from: https://ianlondon.github.io/blog/encoding-cyclical-features-24hour-time/
    
    Included for testing. Possible weaknesses: THIS double weights
    the time feature; inapplicable for tree models (each feature
    might not be present in split).
    """
    
    seconds_in_day = 24*60*60

    df['hr_sin'] = np.sin(2*np.pi*df.PLANNEDTIME_ARR/seconds_in_day)
    df['hr_cos'] = np.cos(2*np.pi*df.PLANNEDTIME_ARR/seconds_in_day)
    
    return df
    
    
```

```python
def centre_scale(df, route_number):
    
    """
    Centres and scales the continuous features' values
    Input: dataframe
    Output: dataframe with its continuous features scaled
    """
    
    # separate df into non/categorical frames
    
    continuous_subset = df.select_dtypes(include=[np.number])
    noncontinuous_subset = df.select_dtypes(['category'])
    
    # fit a scaler to the continuous features
    # transform the continuous features
    
    standard_scaler = StandardScaler()
    scaled = standard_scaler.fit_transform(continuous_subset)
    scaled_df = pd.DataFrame(scaled, columns=continuous_subset.columns)
    
    # reset indices for rejoining
    
    scaled_df.reset_index(drop=True, inplace=True)
    noncontinuous_subset.reset_index(drop=True, inplace=True)

    # rejoin the parts of the original df
    
    join = pd.concat([scaled_df, noncontinuous_subset], axis=1)
    join.reset_index(drop=True, inplace=True)
    
    # specify scaler name and save location
    
    file_path =\
    '/Users/davidodwyer/Documents/studyCS/Semester_3/data/dataframes/scalers'
    scaler_name = 'scaler_route_' + route_number + '.joblib'
    dir_file = os.path.join(file_path, scaler_name)
    
    # save the scaler object for this route
    
    dump(standard_scaler, dir_file)

    return join
```

```python
# Define a function that will make dummies of a categorical feature or a series
# This will be used repeatedly. Thus DRY code.

def get_dummies(data, cols):
    
    """
    Creates new binary features for each value for each 
    categorical feature -1, or for series
    
    Input: Dataframe, List of columns
    Output: Dataframe with value indicator columes added; or dummified Series
    """
    
    if isinstance(data, pd.DataFrame):
        df = data
        categorical_features = cols
        for feature in categorical_features:
            dummies = pd.get_dummies(df[feature],\
                            prefix=feature, drop_first=True)
            df = pd.concat([df, dummies], axis=1)
            df.reset_index(drop=True, inplace=True)
        for feature in categorical_features:
            df.drop(feature, axis=1, inplace=True)
        return df
```

```python
# define a function that will append the square of 
# the continuous values in the dataset to that dataset

def apply_polynomial_transformation(df):
    
    """
    A function that applies a polynomial transformation
    (squaring value) to continuous feature values.
    
    Input: Dataframe
    Output: Dataframe with the squared value of each int and 
    float variable value appended to the dataframe
    """
    
    # sub-function to return the square of input
    
    def square(num):
        num *= num
        return num
    
    # get column names for int and float features
    # append "-poly" (i.e. polynomial) to them
    
    columns = df.select_dtypes(['int32', 'int64', 'float64']).columns.values.tolist()
    continuous_columns = [x + '—poly' for x in columns]

    # create a copy of dataframe with int and float features only
    continuous_vars = pd.DataFrame(df.select_dtypes(['int32', 'int64', 'float64']))
    continuous_vars.columns=continuous_columns
  
    # apply the square function to the values in all rows of the copied subset
    continuous_vars = continuous_vars.apply(square)
    
    # reset the index
    continuous_vars.reset_index(drop=True, inplace=True)

    # Append the polynomial features to the original dataframe
    # and return it
    
    join = pd.concat([continuous_vars, df], axis=1)
    join.reset_index(drop=True, inplace=True)
    
    return join
```

```python
def lasso_best_model(train_set, train_target, val_set, val_target):
    
    """
    A function that applies L1 ("LASSO") regularisation to 
    dataframe features, varying alpha to find optimal value.
    
    Prints parameters for best alpha model; and MSE of that
    model for in- and out-sample datasets; and prints feature
    weights for the best model.
    """
    
    features = list(train_set.columns.values)
    
    folds = KFold(n_splits=5, shuffle=True, random_state=42)
    
    param_grid = [{'alpha' : [0.0001, 0.0005, 0.00075,
                             0.001, 0.005, 0.0075, 
                             0.01, 0.05, 0.075,
                             0.1, 0.5, 0.75, 
                             1, 5]}]
    
    grid = GridSearchCV(Lasso(), param_grid=param_grid,
                        cv=folds, scoring='neg_mean_squared_error', 
                        return_train_score=True, n_jobs=-1)
    
    grid.fit(train_set, train_target)
    
    best_model = grid.best_estimator_
    
    best_model_score = np.sqrt(-grid.best_score_)
    
    best_model_coefs = list(best_model.coef_.ravel())
    
    features_coefs = list(zip(features, best_model_coefs))
    
    print(best_model)
    print('*'*30)
    print(best_model_score)
    print('*'*30)
    for row in features_coefs:
        print('{0:45} {1}'.format(row[0], row[1]))

    
    out_sample_predictions = best_model.predict(val_set)
    
    out_sample_mse = mean_squared_error(y_pred=out_sample_predictions,\
                                       y_true=val_target)
    
    print("MSE on Validation Set: ", out_sample_mse)
    
    return out_sample_predictions

    
    
```

```python
def rf_best_model(train_set, train_target, val_set, val_target):
    
    """
    Function performs a cross-validated grid search to determine
    best Random Forest Regressor parameters. Varies parameters for
    number of estimators, max depth of tree, and minimum samples 
    per leaf.
    """
    
    # define the parameters to test
    
    n_estimator_values = [15, 25, 50]
    
    max_depth_values = [15, 30, 60]
    
    min_samples_leaf_values = [1,2,4]
    
    # define the parameters grid
    
    param_grid = dict(n_estimators=n_estimator_values,
                     max_depth=max_depth_values, min_samples_leaf =
                     min_samples_leaf_values, max_features = [None])
    
    # initialise grid search object
    
    grid = GridSearchCV(RandomForestRegressor(),param_grid=param_grid,
                        cv=3, scoring='neg_mean_squared_error',
                        n_jobs=-1)
    
    # fit grid search
    
    grid.fit(train_set, train_target)
    
    # extract attributes from fitted grid search
    
    best_rf = grid.best_estimator_
    score = np.sqrt(-grid.best_score_)
    best_params = grid.best_params_
    
    # format feature name and feature importances for printing
    
    feature_importances = best_rf.feature_importances_.tolist()
    features = list(train_set.columns.values)
    zipped_nbs_featuers = list(zip(features, feature_importances))
    
    # print the result of the grid search
    
    print('Best model parameters:')
    print(best_rf)
    print("*"*40)
    print('Best score:', score)
    print("*"*40)
    
    print("Best parameters:")
    for k, v in best_params.items():
        print(k, '—', v)     
    print('*'*40)
    
    print('Feature Importances')
    for row in zipped_nbs_featuers:
        print('{0:45} {1}'.format(row[0], row[1]))
        
    
    # generate prediction using the best model parameters
    # print the validation set Mean Square Error score
        
    out_of_sample_predictions = best_rf.predict(val_set)
    
    out_sample_mse = mean_squared_error(y_pred=out_of_sample_predictions,\
                                       y_true=val_target)
    
    print('*'*40)
    print("Validation Set MSE:", out_sample_mse)
    
    return out_of_sample_predictions
    
    
```

**Point Biserial Correlation**

<br>

Taking a single continuous feature at a time, the strength of its linear relationship with the target values is examined. Point Biserial Correlation is used from Scipy's Stats package to determine the Pearson product-moment correlation coefficient ("r-value") and its associated "p-value", an indicator of whether or not a correlation may be counted as statistically legitimate.

<br>

R-value falls within the bound of -1 and 1, where 0 means that there is no relationship between the feature and dependent variable's values; and 1 means that there is a perfect positive correlation; and -1 means that there is a perfect negative correlation.

<br>

P-value has a significance threshold of 0.05, meaning that above this above a given correlation coefficient cannot be said to be legitiamte.

<br>

```python
# for each continuous feature print the coefficient
# of its correlation with the target feature

for col in X_train.select_dtypes(include=[np.number]).columns.tolist():
    print(col)
    print(stats.pointbiserialr(X_train[col], y_train))
    print('_'*40)
```

Between Categorical & Target 

<br>

<br>

Examine the correlation between categorical features and the target feature using the small multiples approach outlined in Kelleher, MacNamee, D'Arcy (2015) §3.5.1.3.

<br>

If the plots show differing central tendencies and variations – amongst themselves, and, separately, in comparison to the distribution of the target – then a relationship may exist between that feature and the target variable.

<br>

```python
# distribution of target

plt.subplots(figsize=(12,8))
plt.xlim(0, y_train.value_counts().max())
sns.distplot(y_train.value_counts(sort=False), bins=100, kde=True, hist=False)
```

```python
def small_multiples(feature):
    
    total_train = X_train.copy()
    total_train['ACTUALTIME_ARR'] = y_train
    
    feature_values = sorted(X_train[feature].unique().tolist())
    
    for f in feature_values:
        plt.xlim(0, y_train.value_counts().max())
        values = total_train.loc[total_train[feature] == f, 'ACTUALTIME_ARR'].value_counts(sort=False)
        sns.distplot(values, bins=100, kde=True, hist=False)
```

REMOVING THE WEAKEST FEATURE (Time-Period)

<br>

Below we define a function for greater flexibility in testing parameters. Input depth and Number Estimators; dumps model for manual size checking on disk.

<br>

```python
def test_rf_params(depth, n_est):
    
    rf_model = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=depth,
       max_features=None, max_leaf_nodes=None,
       min_impurity_decrease=0.0, min_impurity_split=None,
       min_samples_leaf=4, min_samples_split=2,
       min_weight_fraction_leaf=0.0, n_estimators=n_est, n_jobs=-1,
       oob_score=False, random_state=3, verbose=0, warm_start=False)
    
    
    temp_X_train = X_train_rf.copy()
    temp_X_train.drop('time_period_label', axis=1, inplace=True)
    
    temp_X_test = X_test_rf.copy()
    temp_X_test.drop('time_period_label', axis=1, inplace=True)
    
    rf_model.fit(temp_X_train, y_train)
    
    predictions = rf_model.predict(temp_X_test)
    
    mse = mean_squared_error(y_pred=predictions, y_true=y_test)
    
    mae = mean_absolute_error(y_pred=predictions, y_true=y_test)
    
    feature_importances = rf_model.feature_importances_.tolist()
    features = temp_X_train.columns.tolist()
    zipped_nbs_features = list(zip(features, feature_importances))
    
    print("nEstimators of", n_est, "depth of", depth, "MSE Score:", mse)
    print("nEstimators of", n_est, "depth of", depth, "MAE Score:", mae)
    print('_'*30)
    print("\nFeat. — Importance")
    for row in zipped_nbs_features:
        print('{0:45} {1}'.format(row[0], row[1]))
        
    file_path = '/Users/davidodwyer/Desktop/rf_test_depth' + str(depth) + 'nEst' + str(n_est) + '.joblib'
    
#     with open(file_path, 'wb') as f:
#         pickle.dump(rf_model, f, protocol=pickle.HIGHEST_PROTOCOL)

    joblib.dump(rf_model, file_path, compress=True)
        
    print('_'*50)
    print('\n\n')
```
