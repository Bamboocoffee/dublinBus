```python
path = \
"/Users/davidodwyer/Documents/studyCS/Semester_3/data/original_raw_data/google_transit_dublinbus/trips.txt"
```

```python
trips = pd.read_csv(path)
```

```python
trips.head()
```

```python
route_ids = trips.drop_duplicates(subset='route_id', keep='first').route_id.tolist()
```

```python
route_trip_dict = {}
```

```python
trips.loc[trips.route_id == '60-7-b12-1', 'trip_id'].tolist()
```

```python
# create lookup: for each route_id, find all trips

for route in route_ids:
    temp_list_trips = trips.loc[trips.route_id == route, 'trip_id'].tolist()
    route_trip_dict[route] = temp_list_trips
```

The object is a route > trips_on_route look-up

```python
route_trip_dict
```

```python
# the number of trips per route

print("ROUTE    NUMBER OF TRIPS")
for key, value in route_trip_dict.items():
    print(key, "          ", len(value))
```
