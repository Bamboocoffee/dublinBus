# TRIPS DATASET: GROUP SHAPES BY ROUTE_ID

```python
shapes_by_trip = trips.groupby('route_id').apply(lambda x: x['shape_id'].unique())
shapes_by_trip_dict = shapes_by_trip.to_dict()
shapes_by_trip_dict
```

# GROUP ROUTE_IDS BY ROUTE SHORT NAMES

```python
ids_by_route_short_name = routes.groupby('route_short_name').apply(lambda x: x['route_id'].unique())
ids_by_route_short_name_dict = ids_by_route_short_name.to_dict()
ids_by_route_short_name_dict
```

# GROUP SHAPE IDS BY ROUTE SHORT NAMES

```python
ids_by_route_short_name_dict.keys()
```

```python
ids_by_route_short_name_dict['1'][0]
```

```python
route_short_shapes = dict()

for key in ids_by_route_short_name_dict:
    route_short_shapes[key] = []
    for i in range(len(ids_by_route_short_name_dict[key]) -1):
        shapes_key = ids_by_route_short_name_dict[key][i]
        # not all trip ids in gtfs routes are in gtfs trips
        if shapes_key in shapes_by_trip_dict:
            for j in range(len(shapes_by_trip_dict[shapes_key]) -1): 
                route_short_shapes[key].append(shapes_by_trip_dict[shapes_key][j]) 
```

```python
route_short_shapes
```

```python
print("Key  Number of Shapes")

for k, v in route_short_shapes.items():
    print(k, "", len(v))
```
