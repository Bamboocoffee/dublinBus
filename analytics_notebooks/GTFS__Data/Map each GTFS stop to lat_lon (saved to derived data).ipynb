## Could do more above. For now, try to map each bus stop to a lat / lon

```python
# make sub-table of unique stop ids

merged_unique = merged.drop_duplicates('stop_id', keep='first')
```

```python
merged_unique.shape
```

```python
merged_unique.head(3)
```

```python
merged_unique.loc[merged_unique.stop_name == 'Pearse Street']
```

```python
stop_geo_dict = {}

rows = merged_unique.index

for row in rows:
    stop_geo = merged_unique.stop_id[row]
    lat = merged_unique.stop_lat[row]
    lon = merged_unique.stop_lon[row]
    name = merged_unique.stop_name[row]
    
    if stop_geo not in stop_geo_dict:
        stop_geo_dict[stop_geo] = [lat, lon, name]
        
```

```python
stop_geo_dict
```

```python
with open ('GTFS_stops_geo2.json', 'w', encoding='utf-8') as f:
    json.dump(stop_geo_dict, f, ensure_ascii=False, indent=4)
```
